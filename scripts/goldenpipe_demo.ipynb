## Create Sample Data

Messy data with quality issues and duplicates.

```
import csv

rows = [
    ['name', 'email', 'phone', 'city'],
    ['  John Smith  ', 'JOHN@ACME.COM', '(555) 123-4567', 'new york'],
    ['john smith', 'john@acme.com', '555.123.4567', 'New York'],
    ['Jane Doe', 'jane@corp.com', '555-987-6543', 'Chicago'],
    ['  JANE DOE', 'JANE@CORP.COM', '(555)987-6543', 'chicago'],
    ['Bob Wilson', 'bob@test.com', '555 111 2222', 'Boston'],
]

with open('dirty_data.csv', 'w', newline='') as f:
    csv.writer(f).writerows(rows)

print('Created dirty_data.csv with quality issues and duplicates')
```

## Run the Full Pipeline

GoldenPipe chains: GoldenCheck (scan) → GoldenFlow (fix) → GoldenMatch (dedupe)

```
!goldenpipe run dirty_data.csv
```

## Part of the Golden Suite

- [GoldenCheck](https://github.com/benzsevern/goldencheck) — Validate & profile data quality
- [GoldenFlow](https://github.com/benzsevern/goldenflow) — Transform & standardize data
- [GoldenMatch](https://github.com/benzsevern/goldenmatch) — Deduplicate & match records
- [GoldenPipe](https://github.com/benzsevern/goldenpipe) — Orchestrate the full pipeline
